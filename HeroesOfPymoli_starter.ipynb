```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
summaryplr = pd.DataFrame(index=[0], columns=['Total Players'])
summaryplr['Total Players'] = purchase_data['SN'].nunique()
summaryplr
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
summary = pd.DataFrame(index=[0], columns=['Unique Items' , 'Avg Price', 'Purchase','Total Revenue'])
summary['Unique Items'] = purchase_data['Item ID'].nunique()
summary['Avg Price'] = '$\$${:.2f}'.format(purchase_data['Price'].mean())
summary['Purchase'] = purchase_data['Purchase ID'].count()
summary['Total Revenue'] = '$\$${:.2f}'.format(purchase_data['Price'].sum())


summary

```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
uniquegengroup =  purchase_data.groupby('Gender').nunique()

gender_df = pd.DataFrame()
gender_df['Total Count'] = uniquegengroup["SN"]
gender_df['Percentage'] = (gender_df['Total Count'] / gender_df['Total Count'].sum()) * 100
gender_df['Percentage'] = gender_df["Percentage"].map("{0:.2f}%".format)

gender_df

```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Calculates totals, counts, and unique counts off the purchase data
gentotal = purchase_data.groupby('Gender').sum()           
gencount = purchase_data.groupby('Gender').count()
genunique = purchase_data.groupby('Gender').nunique()

#Anaylsis slices are built off the sum,count, and nunique dataframes

GenderPurchaseAnalysis = pd.DataFrame()
GenderPurchaseAnalysis['purchase count'] = gencount['Purchase ID']

GenderPurchaseAnalysis['avg purchase'] = gentotal["Price"] / GenderPurchaseAnalysis['purchase count']
GenderPurchaseAnalysis['avg purchase'] = GenderPurchaseAnalysis['avg purchase'].map("$\$${:.2f}".format)

GenderPurchaseAnalysis['avg total'] = gentotal["Price"] / genunique["SN"]
GenderPurchaseAnalysis['avg total'] = GenderPurchaseAnalysis['avg total'].map("$\$${:.2f}".format)


GenderPurchaseAnalysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#establish bins for ages

bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
bin_names = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']


#Categorize the existing players
agedemo = pd.DataFrame()
agedemo['AgeValues'] = purchase_data['Age']
agedemo['Age'] = pd.cut(agedemo['AgeValues'], bins, labels=bin_names)

agedemosum = pd.DataFrame(agedemo.groupby(agedemo['Age'], as_index=True).size())
agedemosum.columns = ['Total Count']


agedemosum['Percentage'] = (agedemosum['Total Count'] / agedemosum['Total Count'].sum()) * 100
agedemosum['Percentage'] = agedemosum["Percentage"].map("{0:.2f}%".format)

agedemosum
```
